```python
sentence = "python is easy and python is powerful"

words = sentence.split()

unique_words = []

for word in words:
    if word not in unique_words:
        unique_words.append(word)         
unique_words_count = len(unique_words) 

unique_words = "{" + ", ".join(f"'{word}'" for word in unique_words) + "}" 

print(f"Unique words count : {unique_words_count}")  
print(f"Unique words : {unique_words}")
```

```python
employees = {"Ravi" : 75000, "Anita" : 68000, "Kiran" : 72000}

employee_name = ""  

highest_salary = 0

for name, salary in employees.items():
    if salary > highest_salary:
        employee_name = name     
        highest_salary = salary  

print(f"{employee_name} - {highest_salary}")
```

```python
numbers = [45, 22, 89, 10, 66]
max_num = numbers[0]
min_num = numbers[0]

for num in numbers:
    if num > max_num:  
        max_num = num  
    if num < min_num: 
        min_num = num   

print(f"List : {numbers}")
print(f"Max : {max_num}" )
print(f"Min : {min_num}" )
```

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:        
        count = count + 1    

print(f"Products above 1000 : {count}")
```

```python
attendance = ["P", "P", "A", "P", "P"]
total = len(attendance)     
count = 0                  

for atten in attendance:
    if atten == "P":     
        count += 1       
atten_percentage = (count/total)*100

print(f"Attendance Percentage : {atten_percentage}")
```

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = []                  
for number in phone_numbers:
    if number not in unique_numbers: 
        unique_numbers.append(number)

unique_phone_numbers = "{" + ", ".join (str(number) for number in unique_numbers) +"}"

print(f"Unique phone numbers : {unique_phone_numbers}")
```
