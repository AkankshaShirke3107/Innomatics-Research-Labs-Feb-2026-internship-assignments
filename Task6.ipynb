```python
# Problem 1: Smart Parking Lot Management System

def smart_parking_system(capacity, vehicle_logs):
    parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1
        
        if parked > peak_usage:
            peak_usage = parked

    if parked > capacity:
        status = "Over Capacity - Alert!"
    elif parked == capacity:
        status = "Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Usage:", peak_usage)
    print("Parking Status:", status)


capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def delivery_time_estimator(distance, traffic, weather):
    base_time = distance * 5

    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "Low":
        base_time += 5

    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


delivery_time_estimator(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print("Occupancy:", round(occupancy, 2), "%")
    print("Show Status:", status)


total_seats = 200
booked_seats = [1] * 150

seat_occupancy(total_seats, booked_seats)
```

```python
# Problem 4: Cloud Server Load Classification System

def server_load_classification(cpu_readings):
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    if avg_cpu < 50:
        status = "Normal"
    elif 50 <= avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", round(avg_cpu, 2), "%")
    print("Server Status:", status)


server_load_classification([45, 60, 70, 85, 90])
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def classroom_resource_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", alert)


usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def event_registration_controller(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


event_registration_controller(100, 105)
```
